## 7.4.3 Ejercicio

Realiza dos plots de paralelas de la base de datos iris, uno divido por especie y otro donde el color vaya a la variable especie. ¿Qué puedes decir al respecto?

La flores de la especie `Setosa` tiene un comportamiento notoriamente diferente a las otras, en el sentido de que el ancho y largo del sépalo están correlacionadas positivamente mientras que el largo del pétalo y el ancho del sépalo están en una correlación negativa. En cambio, el comportamiento de las especies `Versicolor` y `Virginica` tiene una correlación del ancho y largo del sépalo están en una correlación negativa, mientras el largo del pétalo y el ancho del sépalo están en una correlación positiva.

```r
library(lattice)
data(iris)
```

```r
colnames(iris)
```

```r
parallelplot(
    ~iris|iris$Species,
    main="Plots de Paralelas por Especies",
)
```

```r
parallelplot(
    ~iris,
    col = iris$Species,
    main="Plot de Paralelas por Especies",
)
```
